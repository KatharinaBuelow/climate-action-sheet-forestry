```python
# Listen der GCM Namen erstellen:

GCM_name_allk = [
    "ACCESS1-0_r1i1p1",
    "ACCESS1-3_r1i1p1",
    "CESM1-BGC_r1i1p1",
    "CMCC-CMS_r1i1p1",
    "CMCC-CM_r1i1p1",
    "HadGEM2-CC_r1i1p1",
    "IPSL-CM5B-LR_r1i1p1",
    "inmcm4_r1i1p1",
]

GCM_name_all = [
    "BNU-ESM_r1i1p1",
    "CCSM4_r1i1p1",
    "CSIRO-Mk3-6-0_r1i1p1",
    "GFDL-ESM2G_r1i1p1",
    "CanESM2_r1i1p1",
    "GFDL-ESM2M_r1i1p1",
    "IPSL-CM5A-LR_r1i1p1",
    "IPSL-CM5A-MR_r1i1p1",
    "MIROC-ESM-CHEM_r1i1p1",
    "MIROC-ESM_r1i1p1",
    "MIROC5_r1i1p1",
    "MPI-ESM-MR_r1i1p1",
    "MRI-CGCM3_r1i1p1",
    "bcc-csm1-1-m_r1i1p1",
    "bcc-csm1-1_r1i1p1",
]
```

```python
def read_url(gcm, s):
    url = (
        "https://github.com/IPCC-WG1/Atlas/raw/main/datasets-aggregated-regionally/data/CMIP5/CMIP5_tas_landsea/CMIP5_"
        + gcm
        + "_"
        + s
        + ".csv"
    )
    print(url)
    csv_file = pooch.retrieve(url=url, known_hash=None)
    df = pd.read_csv(csv_file, index_col=0, header=15)
    return df
```

```python
def yearmean(df, where):
    df["date"] = pd.to_datetime(df["date"])
    dfy = df.groupby(df["date"].dt.year)[where].mean()

    t = dfy.to_frame()
    dfy = t.rename(
        columns={"world": "temperature"}
    )  # nicht mehr individueller Name sondern einfach gcm
    return dfy
```

```python
def deviation(df):
    # Abweichung vom vorindustriellen Zeitraum berechnen

    sel = df.query("date >= 1850 and date <= 1900")
    # Mittel unabhängig von NaN
    PI_mean = sel["temperature"].mean(skipna=True)  # wie im Atlas

    dfn = df.copy()
    dfn["temperature"] = dfn["temperature"] - PI_mean
    
    return dfn
```

```python
def runningmean(df):
    # Berechnung vom laufenden Mittelwert:

    df_running = df.rolling(30, center=True).mean()
    return df_running
```

```python
# Listen Erstellen:
GCM_name_rcp85 = [
    "MPI-ESM-LR_r1i1p1",
    "HadGEM2-ES_r1i1p1",
    "EC-EARTH_r12i1p1",
    "CNRM-CM5_r1i1p1",
    "NorESM1-M_r1i1p1",
    "IPSL-CM5A-MR_r1i1p1",
    "MIROC5_r1i1p1",
    "CanESM2_r1i1p1",
]
GCM_name_rcp45 = [
    "HadGEM2-ES_r1i1p1",
    "EC-EARTH_r12i1p1",
    "CNRM-CM5_r1i1p1",
    "NorESM1-M_r1i1p1",
    "MPI-ESM-LR_r1i1p1",
    "IPSL-CM5A-MR_r1i1p1",
]

GCM_name_rcp26 = [
    "MPI-ESM-LR_r1i1p1",
    "HadGEM2-ES_r1i1p1",
    "EC-EARTH_r12i1p1",
    "CNRM-CM5_r1i1p1",
    "NorESM1-M_r1i1p1",
    "MIROC5_r1i1p1",
    "IPSL-CM5A-LR_r1i1p1",
    "GFDL-ESM2G_r1i1p1",
]
```

```python
def read_GCM_file(where):
    'GCM Daten einlesen'
    df = pd.DataFrame()

    sce = ["rcp85", "rcp45", "rcp26"]

    for s in sce:
        # I only read the gcm-file which have been used in EURO-CORDEX as forcing
        if s == "rcp26":
            liste = GCM_name_rcp26
        elif s == "rcp45":
            liste = GCM_name_rcp45
        else:
            liste = GCM_name_rcp85

        for gcm in liste:
            # read input
            dfs = read_url(gcm, s)
            dfh = read_url(gcm, "historical")

            dft = pd.concat([dfs, dfh])
            dft = dft.reset_index()

            # calculate year mean
            dfy = yearmean(dft, where)

            # calculate deviation from pre-industrial
            dfd = deviation(dfy)

            # calculate running mean
            rdf = runningmean(dfd)

            # add more information for easier plotting:
            rdf["gcm"] = gcm
            rdf["scenario"] = s

            df = pd.concat([df, rdf], axis=0)

    df = df.reset_index()
    return df
```

```python
def read_GCM_file_all(where):
    'GCM Daten einlesen'
    df = pd.DataFrame()

    sce = ["rcp85", "rcp45", "rcp26"]
    scek = ["rcp85", "rcp45"]  # , "rcp26"]

    for s in sce:
        # read all which are not in CORDEX

        liste = GCM_name_all
        # print(liste)

        for gcm in liste:
            # read input
            dfs = read_url(gcm, s)
            dfh = read_url(gcm, "historical")

            dft = pd.concat([dfs, dfh])
            dft = dft.reset_index()

            # calculate year mean
            dfy = yearmean(dft, where)

            # calculate deviation from pre-industrial
            dfd = deviation(dfy)

            # calculate running mean
            rdf = runningmean(dfd)

            # add more information for easier plotting:
            rdf["gcm"] = gcm
            rdf["scenario"] = s

            df = pd.concat([df, rdf], axis=0)

    for s in scek:
        # read all which are not in CORDEX

        liste = GCM_name_allk
        # print(liste)

        for gcm in liste:
            # read input
            dfs = read_url(gcm, s)
            dfh = read_url(gcm, "historical")

            dft = pd.concat([dfs, dfh])
            dft = dft.reset_index()

            # calculate year mean
            dfy = yearmean(dft, where)

            # calculate deviation from pre-industrial
            dfd = deviation(dfy)

            # calculate running mean
            rdf = runningmean(dfd)

            # add more information for easier plotting:
            rdf["gcm"] = gcm
            rdf["scenario"] = s

            df = pd.concat([df, rdf], axis=0)
    df = df.reset_index()
    return df
```

```python
def save(df, out):
    'Daten in Ausgabe Datei speichern'
    print("Output will be in =", out)
    df.to_csv(out, na_rep="NaN")
```

```python
# GCMs welche für EURO-CORDEX als Antrieb verwendet wurden 
# Name der Ausgabedatei:

name = "df_ipcc_temperature_deviation_PI_30y_runningmean_world.csv"
workdir = os.getcwd()
out = os.path.join(workdir, name)

# DataFrame einlesen und Region auswählen:
# Hier wurde 'world' gewählt, es könnte auch eine ander Region sein ("NEU","WCE","MED".....)

df = read_GCM_file("world")
save(df, out)
```

```python
# Alle GCMs 
# Name der Ausgabedatei:
name = "df_ipcc_temperature_deviation_PI_30y_runningmean_world_all.csv"
out = os.path.join(workdir, name)

# DataFrame einlesen und Region auswählen:
# In diesem Fall wurde 'world' gewählt, es könnte auch eine ander Region sein ("NEU","WCE","MED".....)

dfa = read_GCM_file_all("world")
save(dfa, out)
```
