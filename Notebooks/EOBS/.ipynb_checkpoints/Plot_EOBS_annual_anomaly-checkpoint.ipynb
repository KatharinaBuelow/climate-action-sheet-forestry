```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.colors import ListedColormap
```

```python
# organising some directories and output files

# Region
# region = "08212_08215_07334_Region_um_Karlsruhe"
# ort='Karlsruhe'
region = "03358_Heidekreis"
ort = "Landkreis Heidekreis"
# region="03353_Harburg"
# ort= "Landkreis Harburg"
# region="03159_Göttingen"
# ort="Göttingen"


workdir = os.getcwd()
inputdir = "/work/ch0636/eddy/pool/climate_fact_data/EOBS/EUR-11/GERICS/v0.4/GWL_ref1971-2000/spavg_NUTS3_merged/year/"
filenameT = "TG/climdexEOBS_TG_year_weighted_means.csv"
filenameNS = "RRm/climdexEOBS_RRm_year_weighted_means.csv"
inputT = os.path.join(inputdir, filenameT)
inputNS = os.path.join(inputdir, filenameNS)

plotdir = "EOBS-Plots-v"
outdir = os.path.join(workdir.replace("EOBS", plotdir))
if not os.path.exists(outdir):
    os.makedirs(outdir)

plotname = "EOBS_Annual_anomaly_deviation_to_1971_2000_" + region + ".png"
figure = os.path.join(outdir, plotname)
```

```python
def select_region(df, region):
    return df.loc[(df["region"].isin([region]))]
```

```python
def deviation(df, variable):
    df["time"] = pd.to_datetime(df["time"])
    df_ = df.copy()
    start = "1971-01-01"
    end = "2000-12-31"

    mask = (df_["time"] >= start) & (df_["time"] <= end)

    df_new = df_[mask]
    hist = df_new["value"].mean()
    # print(hist)
    if variable != "precipitation":
        # print(df['value']-hist)
        df["Temperature anomaly [°C]"] = df["value"] - hist
    else:
        df["Precipitation anomaly [mm]"] = df["value"] - hist
        df["Precipitation change [%]"] = ((df["value"] - hist) / hist) * 100

    # just nice to have for plotting:
    df["year"] = pd.DatetimeIndex(df["time"]).year

    return df
```

```python
dfT = pd.read_csv(inputT)
dfTr = select_region(dfT, region)
dfTn = deviation(dfTr, "temperature")

dfNS = pd.read_csv(inputNS)
dfNSr = select_region(dfNS, region)
dfNSn = deviation(dfNSr, "precipitation")
```
